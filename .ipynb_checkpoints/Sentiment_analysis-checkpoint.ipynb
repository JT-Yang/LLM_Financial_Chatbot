```python
# define functions 
from bs4 import BeautifulSoup

def strip_html_tags(text):
  soup = BeautifulSoup(text, "html.parser")
  [s.extract() for s in soup(['iframe', 'script'])]
  stripped_text = soup.get_text()
  stripped_text = re.sub(r'[\r|\n|\r\n]+', '\n', stripped_text)
  return stripped_text

def remove_accented_chars(text):
  text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
  return text

def stopwords_removal(words):
    list_stopwords = nltk.corpus.stopwords.words('english')
    return [word for word in words if word not in list_stopwords]
```

```python
#  NLP data cleaning process
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
import tqdm
import unicodedata
import contractions

from nltk.tokenize import word_tokenize

def pre_process_corpus(docs):
  norm_docs = []
  for doc in tqdm.tqdm(docs):
    #case folding
    doc = doc.lower()
    #remove special characters\whitespaces
    doc = strip_html_tags(doc)
    doc = doc.translate(doc.maketrans("\n\t\r", "   "))
    doc = remove_accented_chars(doc)
    doc = contractions.fix(doc)
    doc = re.sub(r'[^a-zA-Z0-9\s]', '', doc, re.I|re.A)
    doc = re.sub(' +', ' ', doc)
    doc = doc.strip()
    #tokenize
    doc = word_tokenize(doc)
    #filtering
    doc = stopwords_removal(doc)
    norm_docs.append(doc)
  
  norm_docs = [" ".join(word) for word in norm_docs]
  return norm_docs
```

```python
%%time
df.sentence = pre_process_corpus(df.sentence)
```

### Tokenizer/Sequence/Encoding labels
* Tokenize: converts text into a format that a model can understand, essentially breaking down text into smaller units called tokens.
* Sequencing: Converts lists of tokens into lists of integers based on a predefined mapping from tokens to integers.
* Padding: Adds zeros (or other predefined values) to the sequences to ensure they all have the same length, allowing them to be batch-processed by the model.
* Truncation: Cuts sequences to a maximum length to ensure uniformity and manage computational constraints.
* encoding labels: labels (e.g., positive, negative, neutral) are typically categorical. Models do not understand textual or categorical labels directly, so these labels need to be converted into a numerical format.

#### Vectorization
Vectorization is a broader term that refers to the process of converting text into numerical vectors. This process can involve several techniques, each representing text data in a way that machine learning algorithms can understand and process. Common methods of vectorization include:
1. Bag of Words (BoW): Creates a vector for each document (a piece of text), where each element of the vector represents the frequency of a specific word in the document. The vector's length corresponds to the vocabulary size of the dataset. BoW models don't preserve the order of words.
2. TF-IDF (Term Frequency-Inverse Document Frequency): Similar to BoW but adjusts the frequencies based on how common words are across all documents. It helps to give more weight to words that are unique to specific documents.
3. Word Embeddings: Transforms individual words into dense vectors of real numbers. Techniques like Word2Vec, GloVe, and FastText generate these embeddings by learning representations where the context of each word is taken into account. The resulting vectors capture semantic meanings and relationships between words.

#### Sequencing
Sequencing, in the context of NLP, specifically refers to converting a list of tokens (words, characters, or subwords) into a sequence of integers where each integer represents a specific token according to a predetermined mapping. This step is crucial in preparing data for models that require sequential input, such as Recurrent Neural Networks (RNNs) and Transformers. Sequencing ensures that the model can process text in the order it appears, which is vital for understanding context and meaning in sentences.
1. Padding and Truncation: After converting text into sequences of integers, these sequences might not have uniform lengths. Padding and truncation are techniques used to ensure that all sequences fed into a model have the same length, either by adding zeros (padding) or cutting off tokens (truncation) to fit a predetermined size.

#### Key Differences
Scope: Vectorization is a broader concept that includes various methods of converting text into numerical representations. Sequencing is a specific type of vectorization focused on preserving the order of tokens in a text.

Use Cases: Vectorization methods like BoW and TF-IDF are often used in models where the order of words is less important, such as traditional machine learning algorithms for classification. Sequencing and word embeddings are crucial for models that rely on the sequence of the input, such as RNNs, LSTMs, and Transformers, which are used for tasks requiring understanding of context, syntax, and semantics.

```python
#tokenize 
from tensorflow.keras.preprocessing.text import Tokenizer

token = Tokenizer()
token.fit_on_texts(X_train)
```

```python
vocab = len(token.index_word) + 1
print("Vocabulary size={}".format(len(token.word_index)))
print("Number of Documents={}".format(token.document_count))
```

* It calculates and reports the total vocabulary size (including an extra token for padding or out-of-vocabulary words).
* It prints out the number of unique words found in the text data.
* It reports the total number of documents processed.

```python
#sequence
X_train = token.texts_to_sequences(X_train)
X_test = token.texts_to_sequences(X_test)
```

```python
%matplotlib inline

train_lens = [len(s) for s in X_train]
test_lens = [len(s) for s in X_test]

fig, ax = plt.subplots(1,2, figsize=(12, 6))
h1 = ax[0].hist(train_lens)
h2 = ax[1].hist(test_lens)
```

```python
#padding

from tensorflow.keras.preprocessing.sequence import pad_sequences

MAX_SEQUENCE_LENGTH = 30
X_train = pad_sequences(X_train, maxlen=MAX_SEQUENCE_LENGTH, padding="post")
X_test = pad_sequences(X_test, maxlen=MAX_SEQUENCE_LENGTH, padding="post")
X_train.shape, X_test.shape
```

```python
# encoding labels (can do this at the beginning)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
num_classes=2

y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

### Build Model

```python
import tensorflow as tf 

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, Activation, Dropout
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GlobalMaxPooling1D 

vec_size = 300
model = Sequential()
model.add(Embedding(vocab, vec_size, input_length=MAX_SEQUENCE_LENGTH)) #embedding layer indicates it's designed for sequatial text data
model.add(Conv1D(64,8, activation="relu"))
model.add(MaxPooling1D(2))
model.add(Dropout(0.1))

model.add(Dense(8, activation="relu"))
model.add(Dropout(0.1))
model.add(Dense(4, activation='relu'))
model.add(Dropout(0.1))       
#Dropout after convolutional and dense layers to reduce overfitting by randomly setting a fraction of input 
#units to 0 at each update during training time.
model.add(GlobalMaxPooling1D())
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=tf.optimizers.Adam(learning_rate=0.0001), metrics=['accuracy'])
model.summary()
```

### Train Model

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint

epochs = 100
batch_size = 4

es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=5)
mc = ModelCheckpoint('./best_model/best_model_cnn1d.h5', monitor='val_accuracy', 
                     mode='max', verbose=1, save_best_only=True)
history = model.fit(X_train, y_train,  batch_size=batch_size, shuffle=True, validation_split=0.1, 
                    epochs=epochs, verbose=1, callbacks=[es, mc])
```

## Evaluation

### Accuracy

```python
from keras.models import load_model

saved_model = load_model('./best_model/best_model_cnn1d.h5')
train_acc = saved_model.evaluate(X_train, y_train, verbose=1)
test_acc = saved_model.evaluate(X_test, y_test, verbose=1)
print('Train: %.2f%%, Test: %.2f%%' % (train_acc[1]*100, test_acc[1]*100))
```

### Identify overfitting

```python
# summarize history for accuracy and loss
# Setting up the figure and subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 10)) # 2 Rows, 1 Column

# Plotting model accuracy on the first subplot
axs[0].plot(history.history['accuracy'], label='Train')
axs[0].plot(history.history['val_accuracy'], label='Test')
axs[0].set_title('Model Accuracy')
axs[0].set_ylabel('Accuracy')
axs[0].set_xlabel('Epoch')
axs[0].legend(loc='upper left')

# Plotting model loss on the second subplot
axs[1].plot(history.history['loss'], label='Train')
axs[1].plot(history.history['val_loss'], label='Test')
axs[1].set_title('Model Loss')
axs[1].set_ylabel('Loss')
axs[1].set_xlabel('Epoch')
axs[1].legend(loc='upper left')

# Display the plots
plt.tight_layout()
plt.show()
```

If the training accuracy continues to increase and reaches very high levels, but the validation accuracy plateaus or even decreases, this divergence suggests the model is overfitting. It implies the model performs exceptionally well on the training data but fails to generalize to new, unseen data.

A well-generalizing model will show a decrease in both training and validation losses until they stabilize.If the training loss keeps decreasing towards zero but the validation loss starts to increase, this is a clear sign of overfitting. The model is learning the training data, including its noise and outliers, rather than capturing the underlying pattern that generalizes well to both seen and unseen data.

From the graphs abovw, we can see the model is not that good

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def predictions(x):
    prediction_probs = model.predict(x)
    predictions = [1 if prob > 0.5 else 0 for prob in prediction_probs]
    return predictions

labels = ['positive', 'negative']
print("CNN 1D Accuracy: %.2f%%" % (accuracy_score(y_test, predictions(X_test))*100))
print("CNN 1D Precision: %.2f%%" % (precision_score(y_test, predictions(X_test), average="macro")*100))
print("CNN 1D Recall: %.2f%%" % (recall_score(y_test, predictions(X_test), average="macro")*100))
print("CNN 1D f1_score: %.2f%%" % (f1_score(y_test, predictions(X_test), average="macro")*100))
print(classification_report(y_test, predictions(X_test)))

cm_df=pd.DataFrame(confusion_matrix(y_test, predictions(X_test)), index=labels, columns=labels)

sns.heatmap(cm_df, annot=True, fmt="g", cmap='Oranges')
plt.title('Confusion Matrix')
plt.ylabel('Actual Labels')
plt.xlabel('Predicted Labels')
plt.show()
```

### ROC AUC

```python
fpr, tpr, thresholds = roc_curve(y_test, predictions(X_test))
plt.plot(fpr, tpr)
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')

print("model AUC score: %.2f%%" % (roc_auc_score(y_test, predictions(X_test))*100))
```

### Deep Learning Models
Recurrent Neural Networks (RNN): RNNs are particularly suited for sequential data like text. They can capture the temporal dynamics of sentences, which is beneficial for understanding the sentiment expressed in texts.

Long Short-Term Memory (LSTM): A special kind of RNN capable of learning long-term dependencies. LSTMs are particularly effective for sentiment analysis because they can remember or forget information over long sequences, capturing the overall sentiment better.

Convolutional Neural Networks (CNN): Though primarily known for image processing, CNNs have been successfully applied to NLP tasks. They can capture spatial hierarchies in data, making them effective for extracting sentiment from textual patterns.

Bidirectional Encoder Representations from Transformers (BERT): BERT and similar transformer-based models like GPT (Generative Pretrained Transformer), RoBERTa, and XLNet have set new standards for NLP tasks, including sentiment analysis. They leverage attention mechanisms to capture contextual relations between words in a text.

Transformers with Transfer Learning: Utilizing pre-trained models like BERT, fine-tuned on specific sentiment analysis datasets, has become a common and powerful approach. These models can capture deep linguistic features and nuances, significantly improving the accuracy of sentiment classification.
